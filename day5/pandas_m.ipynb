clean user_id by replacing empty strings with nan and converting values to uppercase


```python
#(df.replace(NaN,columns).uppercase())
df['user_id']=df['user_id'].replace("",np.nan).str.upper()
df['user_id']
```

convert age to numeric and replace invalid  values (negative or >100) with NaN

```python
df['age']=pd.to_numeric(df['age'],errors='coerce')
df.loc[(df['age']<0) | (df['age']>100),'age']=np.nan
df['age']
```

clean salary by removing currency symbols and converting it to numeric 

```python
df['salary']=(
    df['salary']
    .astype(str)
    .str.replace('₹','',regex=False)
    .replace('N/A',np.nan)
)
df['salary']=pd.to_numeric(df['salary'],errors='coerce')
df['salary']
```

convert join_date into datatime format handling multiple date formats


```python
df['join_date']=pd.to_datetime(df['join_date'],errors='coerce',dayfirst=True)
df['joining_year']=df['join_date'].dt.year
df[['join_date', 'joining_year']]
```

extract joining year from the cleaned join_date

```python
df['join_date']=pd.to_datetime(df['join_date'],format="%d-%m-%y",errors='coerce',dayfirst=True)
df['join_date']
```

```python
df['joining_year']=df['join_date'].dt.year
df['joining_year']
```

standarize department values by trimming spaces and converting to uppercase

```python
df['department']=(
    df['department']
    .str.strip()
    .str.upper()
)
df['department']
```

normalize city names to a consistent case and spelling

```python
df['city']=(
    df['city']
    .str.title()
    .str.strip()
)
df['city']
```

convert performance score into numeric  by mapping text values to numbers

```python
performance_map={'poor': 0, 'excellent': 1}

df['performance_score']=df['performance_score'].replace(performance_map)
df['performance_score']=pd.to_numeric(df['performance_score'],errors='coerce')
df['performance_score']
```

clean experience_years by converting text values to numeric and removing negatives

```python
df['experience_years']=df['experience_years'].replace('five',5)
df['experience_years']=pd.to_numeric(df['experience_years'],errors='coerce')
df.loc[df['experience_years']<0,'experience_years']=np.nan
df.loc[df['experience_years']<0,'experience_years']
```

```python
df['is_active'].value_counts(dropna=False)
```

convert is_active values(yes/no/true/false) into boolean format

```python
active_map = {
    "yes": True,
    "no": False,
    "true": True,
    "false": False,
    "1": True,
    "0": False
}

df['is_active'] = (
    df['is_active']
    .astype(str)
    .str.strip()
    .str.lower()
    .map(active_map)   # <-- IMPORTANT: map instead of replace
)

df['is_active'] = df['is_active'].astype('boolean')
df['is_active']

```

identify and remove duplicate records based on user_id

```python
df=df.drop_duplicates(subset='user_id')
df
```
