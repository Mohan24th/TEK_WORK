```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("Practice-Datasets .csv")
df.head()
```

### 1. Create a histogram to analyze the distribution of Age

```python
plt.figure(figsize=(10,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black",alpha=0.7)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

### 2. Plot a pie chart showing the percentage distribution of Gender

```python
plt.figure(figsize=(10,5))
plt.pie(df["Gender"].value_counts(), labels=df["Gender"].value_counts().index, autopct="%1.1f%%", colors=["lightcoral", "lightskyblue"])
plt.title("Gender Distribution")
plt.axis("equal")   
plt.show()
```

### 3. Create a bar chart showing the number of customers in each City

```python
plt.figure(figsize=(10,5))
plt.bar(df["City"].value_counts().index, df["City"].value_counts().values, color="lightgreen", edgecolor="black")
plt.title("City Distribution")
plt.xlabel("City")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

### 4. Plot a box plot to analyze Monthly Income distribution

```python
plt.figure(figsize=(10,5))
plt.boxplot(df["Monthly_Income"],patch_artist=True, vert=False)
plt.title("Box Plot of Monthly Income")
plt.xlabel("Monthly Income")
plt.grid()
plt.show()
```

### 5. Create a histogram of Spending Score and interpret the skewness.

```python
plt.figure(figsize=(10,5))
plt.hist(df['Spending_Score'], bins=20, color="lightblue", edgecolor="black", alpha=0.7)
plt.title("Distribution of Spending Score")
plt.xlabel("Spending Score")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
print("Skewness:",df['Spending_Score'].skew())
```

### 6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df = df.sort_values('Purchase_Date')

daily_sales = df.groupby('Purchase_Date')['Purchase_Amount'].sum()

plt.figure(figsize=(12,6))

plt.plot(
    daily_sales.index,
    daily_sales.values,
    linewidth=2
)

plt.title('Daily Purchase Trend', fontsize=16, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Total Purchase Amount')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.resample('M', on='Purchase_Date')['Purchase_Amount'].sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, linewidth=4)

plt.title('Monthly Purchase Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

### 7. Create a bar plot to compare average Monthly Income by Gender.

```python
avg_income=df.groupby('Gender')['Monthly_Income'].mean()
plt.figure(figsize=(8,5))
plt.bar(avg_income.index, avg_income.values, color=['lightcoral', 'lightskyblue'], edgecolor='black')
plt.title('Average Monthly Income by Gender', fontsize=16, fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Average Monthly Income')    
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 8. Plot a box plot of Purchase Amount grouped by Product Category

```python
category=df['Product_Category'].unique()
category_sales=df.groupby('Product_Category')['Purchase_Amount'].sum().reindex(category)
plt.figure(figsize=(10,6))
plt.boxplot(category_sales, patch_artist=True)
plt.title('Box Plot of Total Sales by Product Category', fontsize=16, fontweight='bold')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

### 9. Create a pie chart to show Product Category distribution

```python
plt.figure(figsize=(10,5))
plt.pie(df["Product_Category"].value_counts(), labels=df["Product_Category"].value_counts().index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Product Category Distribution")
plt.axis("equal")
plt.show()
```

### 10. Plot a histogram of Purchase Amount with appropriate bin size

```python
plt.figure(figsize=(10,5))
plt.hist(df['Purchase_Amount'], bins=20, color="lightblue", edgecolor="black", alpha=0.7)
plt.title("Distribution of Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

### 11. Create a bar chart showing average Spending Score by City.

```python
average_spending_score=df.groupby('City')['Spending_Score'].mean()
plt.figure(figsize=(10,6))
plt.bar(average_spending_score.index, average_spending_score.values, color="lightcoral", edgecolor="black")
plt.title('Average Spending Score by City', fontsize=16, fontweight='bold')
plt.xlabel('City')
plt.ylabel('Average Spending Score')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 12. Plot a line chart showing total Purchase Amount per month

```python
monthly_sales = df.resample("M", on="Purchase_Date")["Purchase_Amount"].sum()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, linewidth=4)
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Monthly Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 13. Create a box plot comparing Monthly Income across different Cities.

```python
plt.figure(figsize=(10,5))

df.boxplot(column="Monthly_Income", by="City")

plt.suptitle("")   # removes automatic title
plt.title("Monthly Income by City")
plt.xlabel("City")
plt.ylabel("Monthly Income")

plt.xticks(rotation=45)
plt.show()
```

### 14. Plot a histogram to analyze Age distribution by Gender

```python

male_age = df[df["Gender"] == "Male"]["Age"]
female_age = df[df["Gender"] == "Female"]["Age"]

plt.figure(figsize=(10,5))

plt.hist([female_age, male_age],
         bins=15,
         label=["Female", "Male"],
         rwidth=0.8)

plt.title("Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.legend()
plt.show()
```

### 15. Create a bar chart showing total Purchase Amount by Product Category

```python
category_total_sales = df.groupby("Product_Category")["Purchase_Amount"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
plt.bar(category_total_sales.index, category_total_sales.values, color="lightblue", edgecolor="black")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 16. Plot a scatter plot between Monthly Income and Spending Score

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Monthly_Income"], df["Spending_Score"])

plt.title("Monthly Income vs Spending Score")
plt.xlabel("Monthly Income")
plt.ylabel("Spending Score")

plt.show()
```

### 17. Create a line chart to visualize daily Purchase Amount variation

```python
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])

# Set date as index
df.set_index("Purchase_Date", inplace=True)

# Resample monthly
monthly_sales = df["Purchase_Amount"].resample("M").sum()

plt.figure(figsize=(10,5))

plt.plot(monthly_sales.index, monthly_sales.values, marker="o")

plt.title("Monthly Purchase Amount Variation", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Purchase Amount")

plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### 18. Plot a box plot of Spending Score grouped by Gender

```python
plt.figure(figsize=(8,5))

df.boxplot(column="Spending_Score", by="Gender")

plt.suptitle("")   # removes automatic extra title
plt.title("Spending Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Spending Score")

plt.show()
```

### 19. Create a bar chart showing customer count by Product Category and Gender

```python
count_data = df.groupby(["Product_Category", "Gender"]).size().unstack()

# Plot grouped bar chart
count_data.plot(kind="bar", figsize=(10,5))

plt.title("Customer Count by Product Category and Gender")
plt.xlabel("Product Category")
plt.ylabel("Customer Count")

plt.xticks(rotation=45)
plt.legend(title="Gender")

plt.tight_layout()
plt.show()
```
